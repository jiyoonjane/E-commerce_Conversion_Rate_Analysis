```python
import pandas as pd

base_path = '/Users/kimminju/Documents/GitHub/05. Final /raw_data/'  # ✅ 오타 수정한 경로
file_names = ['2019-Oct.csv', '2019-Nov.csv', '2019-Dec.csv', '2020-Jan.csv', '2020-Feb.csv']

df_list = []

for file in file_names:
    full_path = base_path + file
    print(f'불러오는 중: {file}')
    df = pd.read_csv(full_path)
    df_list.append(df)

# 전체 병합
all_events = pd.concat(df_list, ignore_index=True)

# 확인
print(f'총 행 수: {len(all_events)}')
all_events.head()

```

```python
df_all = all_events

# 1. price > 0 인 행 삭제
df_all = df_all[~(df_all['price'] <= 0)].copy()  # <- copy() 추가

# 2. brand의 null 값을 "unknown"으로 대체
df_all.loc[:, 'brand'] = df_all['brand'].fillna('unknown')

# 3. user_session이 null인 행 삭제
df_all = df_all.dropna(subset=['user_session'])

# 4. category_code 컬럼 삭제
df_all = df_all.drop(columns=['category_code'])
```

```python
pd.set_option('float_format', '{:.2f}'.format) #전체 숫자 확인
df_all.info()
df_all.describe()
```
